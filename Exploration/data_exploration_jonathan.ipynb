### Question: What and how long is the longest session?
Answer: Session 54433 includes 100 items

```python
longest_session_id = 0
longest_session_length = 0
current_session_id = 0
current_session_length = 1

for row in df_full_session.itertuples():
    if row[3] == current_session_id:
        current_session_length += 1
    else:
        current_session_id = row[3]
        current_session_length = 1
    if current_session_length > longest_session_length:
        longest_session_length = current_session_length
        longest_session_id = current_session_id
print(longest_session_id, longest_session_length)
```
